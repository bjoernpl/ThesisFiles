# Some plots for thesis visualizations

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.cm as cm
import numpy as np
```

### 1. Activation functions

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def relu(x):
    return max(0.0, x)

def tanh(x):
    return np.tanh(x)
```

```python
xs = np.arange(-3,3, 0.1)
sigmoids = np.vectorize(sigmoid)(xs)
relus = np.vectorize(relu)(xs)
tanhs = np.vectorize(tanh)(xs)
plt.axhline(y=0, color='grey')
plt.axvline(x=0, color='grey')
plt.plot(xs, sigmoids, label = "Sigmoid activation")
plt.plot(xs, relus, label = "ReLU activation")
plt.plot(xs, tanhs, label = "tanh activation")
plt.legend()
#plt.savefig("./plots/activations.pdf")
plt.show()
```

### 2. Surface function for gradient descent visualization

```python
def surface(x,y):
    return  2 / np.exp((x-0.5)**2 + y**2)- 2 / np.exp((x+0.5)**2 + y**2)
```

```python
fig = plt.figure(figsize=(10,10))
xs = np.linspace(-2,2,50)
ys = np.linspace(-2,2,50)
ax = fig.add_subplot(111, projection='3d')
zs = np.zeros((len(xs), len(ys)))
for ix, x in enumerate(xs):
    for iy, y in enumerate(ys):
          zs[ix, iy] = surface(x, y)
            
X, Y = np.meshgrid(xs,ys)
ax.plot_surface(X, Y, zs, rstride=1, cstride=1, cmap="viridis")
plt.tight_layout()
#plt.savefig("./plots/loss_curve.pdf")
plt.show()
```
